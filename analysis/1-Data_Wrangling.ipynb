### Data Cleaning <a href="#top" style="font-size:12px">(back to top)</a>
------

> Now that I have all raw data wrangled, I will clean it to isolate the Least Developed Countries (LDCs) and their data from 2000-2018.

```python
# List of LDCs 

ldc = ['Afghanistan',
       'Angola',
       'Bangladesh',
       'Benin',
       'Bhutan',
       'Burkina Faso',
       'Burundi',
       'Cambodia',
       'Central African Republic', 
       'Chad',
       'Comoros',
       'Congo, Dem. Rep.',
       'Djibouti',
       'Eritrea',
       'Ethiopia', 
       'Gambia',
       'Guinea', 
       'Guinea-Bissau', 
       'Haiti', 
       'Kiribati', 
       'Lao', 
       'Lesotho', 
       'Liberia',
       'Madagascar',
       'Malawi', 
       'Mali', 
       'Mauritania',
       'Mozambique',
       'Myanmar',
       'Nepal',
       'Niger',
       'Rwanda',
       'Sao Tome and Principe',
       'Senegal',
       'Sierra Leone', 
       'Solomon Islands', 
       'Somalia', 
       'Sudan',
       'Tanzania', 
       'Togo',
       'Tuvalu',
       'Uganda',
       'Vanuatu',
       'Yemen',
       'Zambia']
```

```python
len(ldc)
```

> - This [list of LDCs](https://www.un.org/development/desa/dpad/least-developed-country-category.html) was sourced from the United Nations.
> - 2 countries, South Sudan and Timor-Leste, have not been included in this list because they were not classified as a LDC until 2012 and 2003 respectively. 
> - **2021 Update**: *when this analysis was performed, Vanuatu was listed as a LDC by the United Nations. Since that time, Vanuatu has graduated from being considered a LDC.*

```python
# Create dataframe for LDCs from 2000-2018.

df_ldc = df_aid_received.loc[ldc, '2000':'2018']
df_ldc
```

```python
# Any missing values or wrong dtypes?

df_ldc.info()
```

```python
df_ldc[df_ldc.isna().any(axis=1)]
```

> I will go ahead and impute the mean for Eritrea and Sudan.

```python
df_ldc = df_ldc.apply(lambda row: row.fillna(row.mean()), axis=1).round(2)
df_ldc
```

```python
# confirm values have been imputed.

df_ldc.isna().sum().sum()
```

> With missing values taken care of, I want to take a closer look at Tuvalu. At first glance, it looks like Tuvalu receives a disproportionate amount of aid compared to all other countries. I want to find out whether or not Tuvalu is an outlier, and if so, is it worth dropping from `df_ldc`. I'll define an outlier using the [1.5xIQR rule](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/identifying-outliers-iqr-rule#:~:text=A%20commonly%20used%20rule%20says,or%20below%20the%20first%20quartile.). 
>
> First I want to quickly visualize the average aid received by each country from 2000-2018. 

```python
df_ldc['mean'] = df_ldc.mean(axis=1).round(2)
df_ldc
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=df_ldc, x=df_ldc['mean'], y=df_ldc.index, color='royalblue');
```

> Based on this visualization, which plots the mean for each country from 2000-2018, Tuvalu looks like an outlier. However, I'll confirm that with the 1.5xIQR Rule. All this means, is that I am testing to see whether or not Tuvalu's mean lies outside of the following value, *Upper quartile (Q3) + 1.5xIQR*. If it does then I will define Tuvalu as an outlier.

```python
df_ldc['mean'].describe()
```

```python
# Find interquartile range, IQR

q1 = df_ldc['mean'].quantile(0.25)
q3 = df_ldc['mean'].quantile(0.75)
IQR = q3 - q1
IQR
```

```python
# Test Tuvalu against 1.5xIQR rule.

tuvalu_mean = df_ldc.loc['Tuvalu', 'mean']
rule = q3 + 1.5 * IQR

print(tuvalu_mean,',', rule)
tuvalu_mean > rule
```

> Tuvalu can now be defined as an outlier. Tuvalu's mean, *1,782.53*, lies outside the upper limit of the rule, *140.08*. Therefore, I will go ahead and drop Tuvalu from `df_ldc`.
>
> There a few other values that also lie outside of this limit, such as Kiribati, Solomon Islands, Djibouti, Liberia, Sao Tome, and Vanuatu, however we won't drop these countries solely based on how extreme of an outlier Tuvalu is. Since this group consists of mean values that are fairly similar, dropping them may do more harm to this analysis than good.

```python
# drop Tuvalu

df_ldc.drop('Tuvalu', inplace=True)
df_ldc
```
