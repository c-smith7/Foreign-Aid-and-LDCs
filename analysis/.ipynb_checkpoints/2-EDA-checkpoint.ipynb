## Introduction <a href="#top" style="font-size:12px">(back to top)</a>
-----
> The dataset that was chosen for this analysis was the Gapminder dataset. Specifically, I used the following data tables: 
> - Foreign Aid Received per capita (current USD&#36;)
> - Human Devlopment Index (HDI)
> - Gini Coefficient
> - Total Fertility Rate 
>
> I chose these because I was genuinely interested in seeing if there was any correlation between the amount of foreign aid a country receives and economic indicators such as HDI, Gini Coefficient, and Fertility rates. 
>
>I focused all foreign aid anlaysis on a per capita basis, in order to gain a more balanced insight into how the data affects individual countries. 
>
>In addition, I will be analyzing data from 2000 to 2018. The data in this timeframe is the most complete and most recent data available from the Gapminder data set. 
>
>Finally, I chose to do my analysis on a subset of countries called LDCs, or Least Developed Countries. These are countries defined by the UN who have a GNI per capita less than &#36;1,025, in addition to other factors such as the Human Assets Index and the Economic Vulnerability Index. I decided to focus on this subset of countries because their economies are very dependent on foreign aid, when compared to non-LDCs who also receive foreign aid. This is important to my analysis, because it allows me to isolate the dependent varibale, foreign aid, as much as possible. In other words, since I am anlayzing countries who rely more on foreign aid, any correlations found between the amount of foreign aid and other economic indicators, will be more statistically significant.
>
>### Reasearch Question 1
>----
> - Is there a positive correlation between the amount of foreign aid per capita an LDC country receives, and the economic indicators: HDI, Gini coefficient, and fertility rates?
>
>### Research Question 2
>----
> - Does the amount of foreign aid per capita have any impact on the strength of this correlation? In other words, does a country who receives more foreign aid per capita see a stronger correlation with the economic indicators, than a country who receives less foreign aid per capita?

### Data Cleaning 
> Now that I have general dataframes imported, I can shave them down the data that I need to for my analysis. 
>
> First I will isolate the LDC countries and the timeframe 2000-2018. 

```python
# before I create an LDC dataframe, I will create a list variable for LDC countries.

ldc = ['Afghanistan', 'Angola', 'Bangladesh', 'Benin', 'Bhutan','Burkina Faso', 'Burundi', 'Cambodia', 
'Central African Republic', 'Chad', 'Comoros', 'Congo, Dem. Rep.', 'Djibouti', 'Eritrea',
'Ethiopia', 'Gambia', 'Guinea', 'Guinea-Bissau', 'Haiti', 'Kiribati', 'Lao', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi', 
'Mali', 'Mauritania', 'Mozambique', 'Myanmar', 'Nepal', 'Niger', 'Rwanda', 'Sao Tome and Principe', 'Senegal', 'Sierra Leone', 
'Solomon Islands', 'Somalia', 'Sudan', 'Tanzania', 'Togo', 'Uganda', 'Vanuatu', 'Yemen', 'Zambia']
```

```python
# now I can create the LDC dataframe for aid per capita. 
df_ldc = df.loc[ldc, '2000':'2018']
df_ldc.head()
```

```python
# any missing values?
df_ldc[df_ldc.isna().any(axis=1)]
```

```python
# there aren't that many missing values so I will impute w/ mean.

df_ldc = df_ldc.apply(lambda row: row.fillna(row.mean()), axis=1).round(2)
df_ldc.head()
```

```python
# confirm all missing values are filled.
df_ldc.isna().sum().sum()
```

>Now, in order to answer my second research question, I am going to seperate the LDC countries into two groups. One group will consist of countries who receive the most amount of foreign aid per capita on average, and the other will be those countries who receive the least on average. I will define these two groups by finding which countries fall in the upper quartile of aid received, and which countries fall in the lower quartile of aid received. By doing this I will be able to compare these two groups to see how their correlations are affected based on the amount of foreign aid they receive. 

```python
# let's first find the average amount of aid each country has received from 2000 to 2018. 

ldc_means = []

for country in ldc:
    mean_ldc = df_ldc.loc[country].mean()
    ldc_means.append(mean_ldc)
    
ldc_means[:10]
```

```python
# Now let's add the means to the dataframe so that we can .describe()
df_ldc['ldc_means'] = df_ldc.mean(axis=1).round(2)
df_ldc.head()
```

```python
# Now we can find the values that will define our quartiles.
df_ldc.ldc_means.describe()
```

>Now we can make seperate dataframes for countries the fall within the upper quartile and countries that fall within the lower quartile.

```python
df_upper = df_ldc.query('ldc_means >= 77.0825').loc[:, '2000':'2018']
df_upper
```

```python
df_lower = df_ldc.query('ldc_means <= 38.98').loc[:, '2000':'2018']
df_lower
```

> Now I will able to use these dataframes during my exploratory data analysis.

> Next let's clean the HDI data.

```python
# lists of countries in each quartile to use when forming dataframes of other variables. 

countries_upper = df_upper.index.tolist()
countries_lower = df_lower.index.tolist()
countries_upper, countries_lower
```

```python
# hdi of upper quartile countries. 
df_hdi_upper = df_hdi.loc[countries_upper, '2000':'2015']
df_hdi_upper
```

> There are some missing values here that I was able to find couple of these values from a UNDP and OECD report for the years 2000 and 2005. Let's insert those..

```python
df_hdi_upper.loc['Bhutan', '2000'] = 0.494
df_hdi_upper.loc['Bhutan', '2005'] = 0.512
df_hdi_upper.loc['Kiribati', '2000'] = 0.552
df_hdi_upper
```

> I was only able to find 3 values, 2 for Bhutan and 1 for Kiribati, which I can use to interpolate. I will also interpolate Vanuatu to keep things consistent, since using the mean will show an artificial drop in hdi from 2000-2004. 

```python
# Bhutan interpolation, USE FOR LOOP HERE
df_hdi_upper.loc['Bhutan'] = df_hdi_upper.loc['Bhutan'].interpolate().round(3)
df_hdi_upper
```

```python
# Vanuatu interpolation. 
df_hdi_upper.loc['Vanuatu'] = df_hdi_upper.loc['Vanuatu'].interpolate(limit_direction='backward')
df_hdi_upper
```

>Obviously this just repeats the value from 2005, but I figured that this would be less detrimental to the analysis than imputing the mean, because when doing this it shows an artificial drop in hdi from 2000-2004 since the mean is 0.589.

```python
# Kiribati interpolation.
df_hdi_upper.loc['Kiribati'] = df_hdi_upper.loc['Kiribati'].interpolate().round(3)
df_hdi_upper
```

> Ther is our main HDI dataframe cleaned for the upper quartile of countries. Now we need to do the same thing for the lower quartile countries. 

```python
# hdi of lower quartile countries. 
df_hdi_lower = df_hdi.loc[countries_lower, '2000':'2015']
df_hdi_lower
```

>Small problem, which I mentioned above, HDI doesn't have data for 2016-2018. I was able to find the missing data from hdr.undp.org/en/data, which is the official body that produces HDI. Below I will import that csv and then clean it a bit to get the info I need, and then add it to the above dataframes. 

```python
df_hdi_16_18 = pd.read_csv('hdi_2016_2018.csv', header=1)
df_hdi_16_18.head()
```

>Ok so I can clean this df up a bit. I only want the columns: country and 2016-2018. I also only want the rows that correspond to the countries in my upper and lower groups. 

```python
df_hdi_16_18 = df_hdi_16_18[['Country', '2016', '2017', '2018']].set_index('Country')
df_hdi_16_18
```

```python
df_hdi_16_18_upper = df_hdi_16_18.loc[countries_upper]
df_hdi_16_18_upper
```

```python
# the name for Congo is different in 16-18 data, so lets make a new temp. list. 
countries_lower_temp = ['Bangladesh', 'Myanmar', 'Angola', 'Nepal', 'Madagascar', 'Sudan', 'Ethiopia',
                        'Togo', 'Guinea', 'Chad', 'Congo (Democratic Republic of the)']

df_hdi_16_18_lower = df_hdi_16_18.loc[countries_lower_temp]
df_hdi_16_18_lower
```

> OK now we have our data for 2016-2018, now we can concant the data to df_hdi_upper and df_hdi_lower.

```python
# concat 2016-2018 data for upper countries. 
df_hdi_upper = pd.concat([df_hdi_upper, df_hdi_16_18_upper], axis=1)
df_hdi_upper
```

```python
# concat 2016-2018 data for lower countries. 
df_hdi_lower = pd.concat([df_hdi_lower, df_hdi_16_18_lower], axis=1)
df_hdi_lower

# oops, congo has two different names, let's fix that. 
```

```python
# we can shift those 3 values up. 
values = df_hdi_16_18_lower.loc['Congo (Democratic Republic of the)']

df_hdi_lower.loc['Congo, Dem. Rep.'] = df_hdi_lower.loc['Congo, Dem. Rep.'].fillna(values)
df_hdi_lower
```

```python
# and then we can drop the last row. 

df_hdi_lower.drop('Congo (Democratic Republic of the)', axis=0, inplace=True)
df_hdi_lower
```

```python
df_hdi_upper
```

>And now we have the cleaned dataframes of hdi for upper and lower quartile countries. 

>Next let's clean Gini and Fertility rates. 

```python
# gini of upper quartile countries.
df_gini_upper = df_gini.loc[countries_upper, '2000':'2018']
df_gini_upper
```

```python
# gini of lower quartile countries.
df_gini_lower = df_gini.loc[countries_lower, '2000':'2018']
df_gini_lower
```

```python
# fertility rates of upper quartile countries.

df_fertility_upper = df_fertility.loc[countries_upper, '2000':'2018']
df_fertility_upper
```

```python
# fertility rates of lower quartile countries.

df_fertility_lower = df_fertility.loc[countries_lower, '2000':'2018']
df_fertility_lower
```

### Research Questions 1 & 2
> Since, the data has already been seperated into two groups, upper quartile foreign aid per capita and lower quartile foreign aid per capita, I will be able to answer both of my research questions simultaneously. I will use my visualizations to compute correlations, while also comparing these correlations between upper quartile and lower quartile countries. 


> First we can calculate the average amount of aid per capita each group received collectively per year. These will be the x-values we can use in our scatter plots to find our correlations. 

```python
# Average aid received per year by countries in the upper quartile. 
avg_aid_upper = df_upper.mean()
avg_aid_upper
```

```python
# Average aid received per year by countries in the lower quartile. 
avg_aid_lower = df_lower.mean()
avg_aid_lower
```

> Now that we have found the average aid received by each group of LDC countries, those in the upper and lower quartiles, we can now do the same thing with each variable, and then plot these mean values against each other. Let's first look at HDI. 

```python
# average hdi per year for upper countries
avg_hdi_upper = df_hdi_upper.mean().round(3)
avg_hdi_upper
```

>Why is 2016 to 2018 not being calculated? Let's check the dtypes of each column..

```python
# Ok, so 2016-2018 were strings, so let's change those to floats. 

df_hdi_upper = df_hdi_upper.apply(lambda columns: columns.astype(np.float64))
df_hdi_upper.info()
```

>Let's do the same thing for hdi_lower.

```python
df_hdi_lower = df_hdi_lower.apply(lambda columns: columns.astype(np.float64))
df_hdi_lower.info()
```

>Now lets try to find the means again..

```python
# average hdi per year for upper quartile countries, second try. 
avg_hdi_upper = df_hdi_upper.mean().round(3)
avg_hdi_upper
```

```python
# average hdi per year for lower quartile countries.
avg_hdi_lower = df_hdi_lower.mean().round(3)
avg_hdi_lower
```

>There we go!

>Let's call avg_aid values just so they are within sight.

```python
avg_aid_upper, avg_aid_lower
```

>Now that we have our values, I believe that we can start to plot now. My idea is to plot these seperately on a single figure with 2 axes, so that they are easily comparable. And then we can calculate the corrrelation coeficients to see how aid and hdi are realated among the two groups, upper and lower. 

```python
# first let's calculate the corrcoef from each plot, so that we can put it on the plot. 

hdi_corr_upper = avg_aid_upper.corr(avg_hdi_upper, method='pearson')
hdi_corr_lower = avg_aid_lower.corr(avg_hdi_lower, method='pearson')
hdi_corr_upper, hdi_corr_lower
```

```python
from matplotlib.ticker import FormatStrFormatter

sns.set()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(40,22))
sns.regplot(avg_aid_upper, avg_hdi_upper, ax=ax1, color='b')
sns.regplot(avg_aid_lower, avg_hdi_lower, ax=ax2, color='c')
ax1.tick_params(axis='both', which='both', labelsize=26)
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax2.tick_params(axis='both', which='both', labelsize=26)
fig.suptitle('Average Aid Received Per Capita vs. Average HDI', fontsize=50, fontweight='semibold', x=0.52)
plt.figtext(x=0.48, y=0.93, s='2000-2018', fontsize=40)
ax1.text(x=125, y=0.535, s='r = 0.807', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax2.text(x=16, y=0.493, s='r = 0.750', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax1.set_title('Upper Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax2.set_title('Lower Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax1.set_ylabel('Average HDI (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_ylabel('Average HDI (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax1.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30);
```

>Interesting, we see a nice positive correlation between aid received and HDI between both groups, upper and lower quartile countries, which answers our first research question. Although the correlation is slightly stronger among the upper quartile countries, I don't think there is a statistically significant difference between to two, therfore the answer to our second question, at least for HDI, is no. The amount of aid per capita a country receives doesn't seem to have a significant effect on their development. 

>Let's now take a look at the Gini coefficient. 

```python
# average gini per year for upper quartile countries.
avg_gini_upper = df_gini_upper.mean().round(1)
avg_gini_upper
```

```python
# average gini per year for lower quartile countries.
avg_gini_lower = df_gini_lower.mean().round(1)
avg_gini_lower
```

```python
# first let's calculate the corrcoef from each plot, so that we can put it on the plot. 
gini_corr_upper = avg_aid_upper.corr(avg_gini_upper, method='pearson')
gini_corr_lower = avg_aid_lower.corr(avg_gini_lower, method='pearson')
gini_corr_upper, gini_corr_lower
```

```python
# Plot

sns.set()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(40,22))
sns.regplot(avg_aid_upper, avg_gini_upper, ax=ax1, color='b')
sns.regplot(avg_aid_lower, avg_gini_lower, ax=ax2, color='c')
ax1.tick_params(axis='both', which='both', labelsize=26)
ax2.tick_params(axis='both', which='both', labelsize=26)
fig.suptitle('Average Aid Received Per Capita vs. Average Gini Coefficient', fontsize=50, fontweight='semibold', x=0.52)
plt.figtext(x=0.48, y=0.93, s='2000-2018', fontsize=40)
ax1.text(x=125, y=38.85, s='r = -0.807', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax2.text(x=16, y=38.9, s='r = -0.830', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax1.set_title('Upper Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax2.set_title('Lower Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax1.set_ylabel('Average Gini (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_ylabel('Average Gini (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax1.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30);
```

>Here we see a negative correlation, which is what we would expect, since gini represents income inequality. The lower gini is, the better. I would consider this a 'postive' correlation in terms of answering the first research question. Very similar correlations here again, which suggests that being in the upper quartile or lower quartile doesn't have much impact on development. 

>Lastly, let's take a look at Fertility rates. 

```python
# average fertility rate per year for upper quartile countries.
avg_fertility_upper = df_fertility_upper.mean().round(2)
avg_fertility_upper
```

```python
# average fertility rate per year for lower quartile countries.
avg_fertility_lower = df_fertility_lower.mean().round(2)
avg_fertility_lower
```

```python
# calculate the corrcoef for each plot. 
fertility_corr_upper = avg_aid_upper.corr(avg_fertility_upper, method='pearson')
fertility_corr_lower = avg_aid_lower.corr(avg_fertility_lower, method='pearson')
fertility_corr_upper, fertility_corr_lower
```

```python
# Plot

sns.set()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(40,22))
sns.regplot(avg_aid_upper, avg_fertility_upper, ax=ax1, color='b')
sns.regplot(avg_aid_lower, avg_fertility_lower, ax=ax2, color='c')
ax1.tick_params(axis='both', which='both', labelsize=26)
ax2.tick_params(axis='both', which='both', labelsize=26)
fig.suptitle('Average Aid Received Per Capita vs. Average Fertility Rate', fontsize=50, fontweight='semibold', x=0.52)
plt.figtext(x=0.48, y=0.93, s='2000-2018', fontsize=40)
ax1.text(x=125, y=3.91, s='r = -0.818', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax2.text(x=16, y=4.55, s='r = -0.764', fontsize=33, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.5'))
ax1.set_title('Upper Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax2.set_title('Lower Quartile Countries', fontsize=40, fontweight='semibold', pad=15)
ax1.set_ylabel('Average Fertility Rate (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_ylabel('Average Fertility Rate (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax1.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30)
ax2.set_xlabel('Average Aid Received (per year)', fontsize=35, fontweight='semibold', labelpad=30);
```

>And with Fertility rates we see the same kind of correlation as Gini. Negative correlation, which makes sense, since a decreasing fertility rate is associated with higher economic development. Again, the correlations between upper and lower quartiles are a bit more seperated, however not enough to make the claim that receiving more or less foreign aid per capita affects development in a significant way.
